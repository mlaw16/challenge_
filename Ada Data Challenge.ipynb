## Question 1: Which region had the most "Grassland pasture and range" in 1974?

```python
# I used a pivot table to create a dataframe with the values I want to sort through. 
df1 = df.pivot_table(values='Grassland pasture and range', index='Year', columns='Region')
df1
```

```python
# I selected the row of 1974 using the .loc method and sorted the values from max to min. 
df1.loc[1974].sort_values(ascending=False)
```

### Ansewer 1: Region is Mountain (total) as it had 307445 grassland pasture and range in 1974.

## Question 2: How many states had at least 2,000 in the Land in Urban areas column for any year prior to 1987?


```python
# I used a pivot table to create a dataframe with the values I want to sort through.
df2 = df.pivot_table(values='Land in urban areas', index='Region or State', columns='Year')
df2
```

```python
# I dropped columns and rows not needed to clean the data. 
df2.drop([1987,1992,1997,2002,2007,2012],axis=1,inplace=True)
df2.drop(['Appalachian','Corn Belt','Delta States','Lake States','Mountain','Northeast','Northern Plains','Pacific','Southeast','Southern Plains'],inplace=True)
```

```python
# Using a conditional statement, I filtered the data to only show values equals or larger than 2000 and then dropped states with nan values.  
df2[df2>=2000.0].dropna(how='all')
```

### Answer 2: California, Florida, Ohio,  and Texas.

## Question 3: What is the average value of the Cropland used for pasture column among all states within the Pacific and Mountain regions for 1964?


```python
# I used a pivot table to create a dataframe with the values I want to sort through.
df3 = df.pivot_table(values='Cropland used for pasture', index='Year', columns='Region')
df3
```

```python
# I selected both columns and took the mean value. 
df3[['Mountain total','Pacific total']].loc[1964].mean()
```

### Answer 3: Average values for Mountain total and Pacific total in 1964 = 4169

## Question 4: Consider the original 13 colonies of England which became what is now the United States. Of the states that have land within the territory of those 13 colonies, which state made the largest contribution to its region's total Forest-use land in 2012?


```python
# I used a pivot table to create a dataframe with the values I want to sort through.
df4 = df.pivot_table(values='Forest-use land (all)', index='Year', columns='Region or State')
df4
```

## Question 5: Which region has had the largest shift in its land use between 1945 and 2012? Please explain your reasoning, including the data that you used to reach that conclusion.


```python
# I used a pivot table to create a dataframe with the values I want to sort through.Then I dropped unwanted columns. 
df5 = df.pivot_table(values='All special uses of land', index='Year', columns='Region')
df55 = df5.drop(['Appalachian','Corn Belt','Delta States','Lake States','Mountain','Northeast','Northern Plains','Pacific','Southeast','Southern Plains'],axis=1)
```

```python
df55
```

```python
# I used the .std method to get the standard deviation values for all the regions between 1945-2012. The standard devation values represents the shift between values from year to year. 
df55.std().sort_values(ascending=False)
```

### Answer 5: Mountain total with standard deviaiton of 12952.634919.
